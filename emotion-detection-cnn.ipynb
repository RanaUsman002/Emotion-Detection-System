# For images

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Model

# Define image input layer
img_size = (224, 224)
channels = 3
img_shape = (img_size[0], img_size[1], channels)
image_input = Input(shape=img_shape)

# Define image processing layers
conv1 = Conv2D(32, (3, 3), activation='relu')(image_input)
maxpool1 = MaxPooling2D((2, 2))(conv1)
flatten1 = Flatten()(maxpool1)

# Hidden layers
dense1 = Dense(64, activation='relu')(flatten1)
dense2 = Dense(32, activation='relu')(dense1)

# Output layer
class_count = 7 # Number of classes
output = Dense(class_count, activation='softmax')(dense2)

# Create model
model = Model(inputs=image_input, outputs=output)

# Compile model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# Calculate steps_per_epoch and validation_steps
steps_per_epoch = len(train_gen)
validation_steps = len(valid_gen)

# Fit the model with image data only
history = model.fit(train_gen,
                    steps_per_epoch=steps_per_epoch,
                    validation_data=valid_gen,
                    validation_steps=validation_steps,
                    epochs=12)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_gen, steps=len(test_gen))
print("Test accuracy:", test_accuracy)
```
